Compare
=======

In this document, we evaluate the above methods on a real (with and without simulating differences) dataset and compare to external methods.

```python
%matplotlib inline

import toolshed as ts
import itertools as it
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

sns.set(style="white", context="talk")
colors = sns.color_palette("Set1", 8)
pool = ts.pool()

import crystal
import aclust # to create correlated clusters
```

We need a generator that reads data and generates features to sent to aclust

```python
N = 12

def feature_gen(fname, samples, logit=True):         
    for i, d in enumerate(ts.reader(fname, header=True)):
        chrom, pos = d['probe'].split(":")
        values = np.array([float(d[s]) for s in samples])
        yield crystal.Feature(chrom, int(pos), crystal.logit(values) if logit else values)                         
```

Now we need to define our model and read in our covariates. First, we want to test 6 vs 6 (male vs female).

```python
methylation_file = "/drive/450k/norm.beta.txt.gz"
methylation_file = "/drive/450k/beta.subset.txt"
covariates_file = "covariates.csv"
np.random.seed(42)

covs_full = pd.read_csv(covariates_file, index_col=0)

covs_full['id'] = covs_full.index
covs_full['plate'] = covs_full.Basename.map(lambda s: s.split("_")[0])
formula = "methylation ~ age + gender"

F = covs_full.query('gender == "F"')
M = covs_full.query('gender == "M"')

covs = pd.concat((F.ix[np.random.random_integers(0, len(F) - 1, size=N / 2), :],
                  M.ix[np.random.random_integers(0, len(M) - 1, size=N / 2), :]), axis=0)
covs.shape

covs.to_csv('covs.meth.txt', index=False)
```

Our input data is a methylation matrix (rows of probes and columns of samples). Here, we have encoded the position into the row names, but you can also add these on the fly as you stream over the data.

```python
# a function to restart the cluster generation each time.
def get_cluster_iter(samples, methylation_file, min_len=1, logit=True):
    return (c for c in aclust.mclust(feature_gen(methylation_file, samples, logit=logit),
                              max_dist=300, max_skip=0)
            if len(c) >= min_len)
```

Mclust returns an iterator of clusters given an iterator of features. The clusters are have correlated methylation. The clustering is done without knowing the study design so it is unbiased.

Each cluster is a list of features from your own feature_gen()

```python
clusters = list(get_cluster_iter(covs.index, methylation_file))
```

```python
region_fh = open('ex-regions.bed', 'w')

cluster_fh = open('ex-meth.txt', 'w')
cluster_fh.write("\t".join(['probe'] + list(covs['id'])) + "\n")

for cluster in crystal.simulate.simulate_regions(clusters, region_fh, class_order=list(covs['gender']), seed=21):
    crystal.utils.write_cluster(cluster, cluster_fh)
cluster_fh.flush()
```

```python
clusters = list(get_cluster_iter(covs.index, "ex-meth.txt", logit=False))
print sum(len(c) for c in clusters)
```

```python
fig, ax = plt.subplots(1)

formula_base = "methylation ~ gender + age"
formula_cpg = formula_base + " + CpG"
coef = "gender"

clean_methods = ["z-score", "z-score robust", "GEE", "OLS cluster", "bumping", "mixed-model"]



methods, models, kwargs = zip(*[
    (crystal.zscore_cluster, formula,          dict(ax=ax, plot_kwargs=dict(color=colors[0], label=clean_methods[0]))),
    (crystal.zscore_cluster, formula,          dict(ax=ax, plot_kwargs=dict(color=colors[1], label=clean_methods[1]), robust=True)),                                       
    (crystal.gee_cluster, formula_cpg,         dict(ax=ax, plot_kwargs=dict(color=colors[2], label=clean_methods[2]))),
    (crystal.ols_cluster_robust, formula_cpg,  dict(ax=ax, plot_kwargs=dict(color=colors[3], label=clean_methods[3]))),
 #   (crystal.bump_cluster, formula,            dict(ax=ax, plot_kwargs=dict(color=colors[4], label=clean_methods[4]))),
    (crystal.mixed_model_cluster, formula_cpg, dict(ax=ax, plot_kwargs=dict(color=colors[5], label=clean_methods[5]))),
                                       ])

results = [crystal.evaluate.evaluate_regions(
                clusters, "ex-regions.bed", covs, model, coef, fn, pool=pool, kwargs=kw)
                                           for (fn, model, kw) in zip(methods, models, kwargs)]
ax.plot([0, 1], [0, 1], 'k--', alpha=0.5)
ax.set_xlim(0, 0.1)
ax.legend(loc="best")
```

```python
formula_base = "methylation ~ gender + age"
formula_cpg = formula_base + " + CpG"
coef = "gender"

modeled = [list(crystal.model_clusters(clusters, covs, formula, coef, fn, pool=pool, **kwargs))
             for fn, formula, kwargs in [
                         (crystal.zscore_cluster, formula_base, {}),
                         (crystal.zscore_cluster, formula_base, dict(robust=True)),
                         (crystal.gee_cluster, formula_cpg, {}),
                         (crystal.mixed_model_cluster, formula_cpg, {})
           ]]
```

```python
fig, ax = plt.subplots(1)
labels = ['z-score', 'z-score robust', 'gee', 'mixed-model']
for i in range(len(modeled)):
    bediter = list(crystal.evaluate.cluster_bediter(modeled[i]))
    tr, ps = crystal.evaluate.evaluate_modeled_regions(bediter, "ex-regions.bed", ax=ax, label=labels[i])
ax.plot([0, 1], [0, 1], 'k--', zorder=-1)
ax.legend(loc="best")
```
